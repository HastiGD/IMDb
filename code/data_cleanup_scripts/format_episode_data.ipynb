# Format Episode Data

## Load Episode data into df

```python
import pandas

path = "/Users/hastigheibidehnashi/DBMS/IMDb/raw_data/"
file = "title.episode.tsv.gz"
episode_df = pandas.read_csv(path+file, sep="\t", na_values="\\N", nrows = 20)
```

## Convert seasonNumber and episodeNumber fields to int

```python
# convert seasonNumber and episodeNumber to int
episode_df["seasonNumber"] = episode_df["seasonNumber"].astype('Int64')
episode_df["episodeNumber"] = episode_df["episodeNumber"].astype('Int64')
display(episode_df)
```

## Export df to .tsv file

```python
# save formatted data as new tsv file
path = "/Users/hastigheibidehnashi/DBMS/IMDb/IMDb_Database/data/"
file = "episode.tsv"
episode_df.to_csv(path+file, sep="\t")
```
